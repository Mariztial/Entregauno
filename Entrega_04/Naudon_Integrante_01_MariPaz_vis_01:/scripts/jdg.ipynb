```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('impacto_ambiental.csv')
print('Planilla:')
df
```

```python
categorias = df['Categoría ']
materialidad = df['Materialidad']
transporte = df['Transporte']
produccion_bolsas = df['Producción bolsas']
c_utilizada_2017 = df['c/utilizada 2017']
c_utilizada_2018 = df['c/utilizada 2018']
c_utilizada_2019 = df['c/utilizada 2019']
c_utilizada_2020 = df['c/utilizada 2020']
c_utilizada_2021 = df['c/utilizada 2021']
potencial_cambio_climatico = df['Potencial de Cambio Climático (GWP100) [Kg CO2 eq.]']
ecotoxicidad_marina = df['Ecotoxicidad\nmarina [Kg 1,4-\nDBeq.]\n']
potencial_acidificacion = df['Potencial de\nacidificación [Kg\nSO2 eq.]\n\n']
consumo_agua = df['Consumo de\nagua [M^3]']
agotamiento_ozono = df['Agotamiento\nozono [Kg\nCFC-11 eq.]\n']
```

## Gráficos de la Planilla
---

```python
plt.figure(figsize=(12, 6))
bar_width = 0.15

x = range(len(categorias))
x_2017 = x
x_2018 = [i + bar_width for i in x]
x_2019 = [i + 2 * bar_width for i in x]
x_2020 = [i + 3 * bar_width for i in x]
x_2021 = [i + 4 * bar_width for i in x]


plt.bar(x_2017, c_utilizada_2017, width=bar_width, label='2017')
plt.bar(x_2018, c_utilizada_2018, width=bar_width, label='2018')
plt.bar(x_2019, c_utilizada_2019, width=bar_width, label='2019')
plt.bar(x_2020, c_utilizada_2020, width=bar_width, label='2020')
plt.bar(x_2021, c_utilizada_2021, width=bar_width, label='2021')

plt.ylabel('Cantidad de bolsas')
plt.title('Bolsas utilizadas durante 2017-2021')
plt.xticks([i + 2 * bar_width for i in x], categorias)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, potencial_cambio_climatico, width=0.4)
plt.ylabel('Potencial de Cambio Climático (GWP100) [Kg CO2 eq.]')
plt.title('Potencial de Cambio Climático')
plt.xticks(df.index, df['Categoría '])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, ecotoxicidad_marina, width=0.4, color = 'orange')
plt.ylabel('Ecotoxicidad marina [Kg 1,4 DBeq.]')
plt.title('Ecotoxicidad Marina')
plt.xticks(df.index, df['Categoría '])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, potencial_acidificacion, width=0.4, color = 'green')
plt.ylabel('Potencial de Acidificación [Kg SO2 eq.]')
plt.title('Potencial Acidificación')
plt.xticks(df.index, df['Categoría '])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, consumo_agua, width=0.4, color = 'red')
plt.ylabel('Consumo de agua [M^3]')
plt.title('Consumo de Agua')
plt.xticks(df.index, df['Categoría '])
plt.show()
```
